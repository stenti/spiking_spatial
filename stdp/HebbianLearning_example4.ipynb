```python
import matplotlib.pyplot as plt
import nest as sim
import numpy as np
sim.ResetKernel()

sim_len = 3000


srt_tm = 2000.
stp_tm = 8000.
```

```python
def plotSpks(ax,t,s): # define function to plot spikes in a common way
    ax.plot(t,s,'.')
    ax.set_ylabel('Neuron')
    ax.set_xlabel('Time (ms)')
    ax.set_xlim([0,sim_len])
```

```python
sim.SetDefaults("stdp_synapse",{"tau_plus": 15.0})

syn_dict = {"model": "stdp_synapse", "alpha": 0.5,"weight": 150.,'Wmax':500.,'delay': 1.}
```

```python
HD = sim.Create("iaf_psc_alpha",3)
g1 = sim.Create("iaf_psc_alpha",3)
view = sim.Create("iaf_psc_alpha",5, params={"tau_minus": 30.0})
```

```python
# view_current = sim.Create('ac_generator ', 1, params = {'offset': 300.,'amplitude':450.,'frequency':5.,'phase': 0.,'stop':stp_tm})

step = 200.
jump = 1400.
I = 500.

v1 = 400.
v2 = 800.
v3 = 1200.
view_current = sim.Create('step_current_generator', 5, params = [{'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v2,v2+step],'amplitude_values':[I,0.0,I,0.0]}])
sim.Connect(view_current,view,'one_to_one')

v1 = v1+jump
v2 = v2+jump
v3 = v3+jump
view_current2 = sim.Create('step_current_generator', 5, params = [{'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v2,v2+step],'amplitude_values':[I,0.0,I,0.0]}])
sim.Connect(view_current2,view,'one_to_one')
HD_current = sim.Create('step_current_generator', 3, params = [{'amplitude_times':[v1-50,v1+step+50.],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v2-50,v2+step+50.],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v3-50,v3+step+50.],'amplitude_values':[I,0.0]}])
sim.Connect(HD_current,HD,'one_to_one')

v1 = v1+jump
v2 = v2+jump
v3 = v3+jump
view_current3 = sim.Create('step_current_generator', 5, params = [{'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v2,v2+step],'amplitude_values':[I,0.0,I,0.0]}])
sim.Connect(view_current3,view,'one_to_one')

v1 = v1+jump
v2 = v2+jump
v3 = v3+jump
view_current4 = sim.Create('step_current_generator', 5, params = [{'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v2,v2+step],'amplitude_values':[I,0.0,I,0.0]}])
sim.Connect(view_current4,view,'one_to_one')
HD_current2 = sim.Create('step_current_generator', 3, params = [{'amplitude_times':[v1+50,v1+step-100.],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v2+50,v2+step-100.],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v3+50,v3+step-100.],'amplitude_values':[I,0.0]}])
sim.Connect(HD_current2,HD,'one_to_one')
v1 = v1+jump
v2 = v2+jump
v3 = v3+jump
view_current5 = sim.Create('step_current_generator', 5, params = [{'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v2,v2+step,v3,v3+step],'amplitude_values':[I,0.0,I,0.0]},
                                                                {'amplitude_times':[v1,v1+step,v2,v2+step],'amplitude_values':[I,0.0,I,0.0]}])
sim.Connect(view_current5,view,'one_to_one')
HD_current3 = sim.Create('step_current_generator', 3, params = [{'amplitude_times':[v3+100,v3+step-50.],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v1+100,v1+step-50.],'amplitude_values':[I,0.0]},
                                                                {'amplitude_times':[v2+100,v2+step-50.],'amplitude_values':[I,0.0]}])
sim.Connect(HD_current3,HD,'one_to_one')
sim_len = v3+500.
```

```python
sim.Connect(HD,g1, "one_to_one", syn_spec={'weight': 500, 'delay': 0.1})
sim.Connect(view,g1, "all_to_all", syn_dict)
```

```python
HD_spikes = sim.Create("spike_detector", 1, params={"withgid": True,"withtime": True})
g1_spikes = sim.Create("spike_detector", 1, params={"withgid": True,"withtime": True})
view_spikes = sim.Create("spike_detector", 1, params={"withgid": True,"withtime": True})

sim.Connect(HD,HD_spikes)
sim.Connect(g1,g1_spikes)
sim.Connect(view,view_spikes)


# m1 = sim.Create("multimeter",1)
# sim.SetStatus(m1, {"withtime":True, "record_from":["V_m"]})
# m2 = sim.Create("multimeter",1)
# sim.SetStatus(m2, {"withtime":True, "record_from":["V_m"]})
# m3 = sim.Create("multimeter",1)
# sim.SetStatus(m3, {"withtime":True, "record_from":["V_m"]})

# sim.Connect(m1, HD)
# sim.Connect(m2, g1)
# sim.Connect(m3, view)
```

```python
sim.Simulate(sim_len)
```
