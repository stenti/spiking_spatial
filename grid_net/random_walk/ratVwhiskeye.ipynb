script for comparing trajectory of whiskeye vs the rat data used to control whiskeyes position

```python
import matplotlib.pyplot as plt
import nest as sim
import numpy as np
import os
from scipy.io import loadmat
import pandas
```

```python
path = os.path.realpath('data/ratVwhiskeye/rat_trajectory.mat')
ratData = loadmat(path)
rat_t = np.around(np.array(ratData['pos_timeStamps'].flatten())*1000)
rat_x = np.array(ratData['pos_x'].flatten())
rat_y = np.array(ratData['pos_y'].flatten())

rat_vel_x = np.diff(rat_x)
rat_vel_y = np.diff(rat_y)
```

```python
posedata = pandas.read_csv(f'data/ratVwhiskeye/rat_data_test_run_repeat_trim.csv')
pos_t = np.array(np.around(posedata['Time']*1000))
pos_x = np.array(posedata['X']*100)
pos_y = np.array(posedata['Y']*100)

pos_x = pos_x[10:]
pos_y = pos_y[10:]
pos_t = pos_t[10:]

vel_x = np.diff(pos_x)
vel_y = np.diff(pos_y)
```

```python
plt.figure(figsize=(20,20))
plt.plot(pos_x-10,pos_y, label = 'whiskeye trajectory')
plt.plot(rat_x,rat_y, label = 'rat trajectory')
plt.legend()
```

```python
plt.plot(vel_x, label = 'whiskeye velx')
plt.plot(rat_vel_x, label = 'rat trajectory velx')
plt.legend()
plt.figure()
plt.plot(vel_y, label = 'whiskeye trajectory vely')
plt.plot(rat_vel_y, label = 'rat trajectory vely')
plt.legend()
```
